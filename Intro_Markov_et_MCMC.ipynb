<p class="bg-primary" style="padding:1em"><b>QUESTION V.1.1.</b> La fonction suivante génère une suite $(X_t)$ : quelles sont les valeurs prises par cette suite ? Expliquez simplement sa dynamique.</p>

```python
def Markov_intro(t_max, p, init):
    X = np.zeros(t_max)
    X[0] = init
    for time in range(1, t_max):
        if (np.random.random() < p):
            X[time] = X[time-1];
        else:
            X[time] = 1 - X[time-1];
    return X
```

<div class="alert alert-warning"><b>REPONSE A LA QUESTION V.1.1. </b>Le premier élément de la suite (qui est $X_0$) est init. Après, le calcul de l'état $X_{t+1}$ est aléatoire, mais il ne dépend que de l'état $X_t$ et du paramètre p, étant calculée de la manière suivante: $X_{t+1}$ sera égal à $X_t$ avec une probabilité p et égal à $1-X_t$ avec une probabilité $(1-p)$. Donc, la loi de calcul de $X_{t+1}$ à partir de $X_t$ est une Bernoulli de paramètre p. La valeur de l'état $X_{t+1}$ ne dépendant que de $X_t$, la suite s'agît d'une chaîne de Markov.
    
</div>

<div class="alert alert-success"> On dit qu'une suite $(Y_n)$ satisfait le <b>théorème ergodique</b> si la suite $(\bar Y_N)$ des moyennes empirique converge presque sûrement vers une constante, <i>i.e.</i> s'il existe une constante $\mu$ telle que $\displaystyle \bar Y_N = \frac{1}{N} \sum_{k=1}^N Y_k \xrightarrow[N \to \infty]{p.s.} \mu. $ </div>

<p class="bg-primary" style="padding:1em"><b>QUESTION V.1.2.</b> Une suite i.i.d. satisfait-elle le théorème ergodique ? Justifiez votre réponse.</p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION V.1.2.</b> D'après le théorème central limite, si $Y_k$ sont des variables i.i.d d'esperance $\mu$ et de variance $\sigma^2$, alors:

$$lim_{N \to \infty} \frac{\sum_{k=1}^N (Y_k - \mu)}{\sqrt{N} \sigma^2} \mathop{\sim} N(0,1) $$

D'où,

$$\frac{1}{N }lim_{N \to \infty} \sum_{k=1}^N Y_k \mathop{\sim} N(\mu,\sigma^2 / N) $$

On voit bien que pour N suffisamment grand, la variance tend vers zero et donc la suite $(\bar Y_N)$ des moyennes empirique converge presque sûrement vers $\mu$. Donc, une suite de variables aléatoires i.i.d satisfait le théorème ergodique. </div>

<p class="bg-primary" style="padding:1em"><b>QUESTION V.1.3.</b> Utilisez le script suivant pour vérifier que $(X_t)$ satisfait le théorème ergodique. Pour quelles valeurs de $p$ la convergence est-elle la plus rapide ? La condition initiale joue-t-elle un rôle ? Expliquez pourquoi.</p>

```python
t_max = 10**2
val_p = np.arange(.1,1,.2)
nb_p = len(val_p)

fig, ax = plt.subplots()
for i in range(nb_p):
    p = val_p[i]
    traj = Markov_intro(t_max, p, 1)
    ax.plot(range(t_max), np.divide(np.cumsum(traj), range(1,1+len(traj))), label="p="+str(p))
plt.legend()
plt.suptitle("Vérification du théorème ergodique")
plt.xlabel("Temps")
plt.ylabel("Moyenne empirique")
```

<div class="alert alert-warning"><b>REPONSE A LA QUESTION V.1.3.</b>
    
La convergence est plus rapide pour des valeurs de p petites. En effet, la courbe correspondant à p=0.1 est celle qui converge plus vite, et la courbe associé à p=0.9 est celle qui converge plus lentement. Cela a du sens, vue que avec une valeur basse du paramètre p, il est peu probable de rester au même état, et très probable de changer d'état, résultant en une convergence plus accélérée. Par contre, si p est élevé, il est peu probable qu'on change d'état et très probable qu'on reste au même état, résultant en une convergence plus lente.

La condition initiale ne change pas le fait ni qu'on va converger ni qu'on converge plus vite pour p petit, mais elle a un impact sur le temps de convergence. En effet, si on commence à 0 ou 1, on va seulement osciller entre ces deux valeurs, alors que si on commence par exemple à 50, on va osciller entre 50 et -49 avant de finalement arriver à osciller entre 0 et 1.

<a id="MCMC"></a><h1 style="border: 5px solid; text-align: center; padding: 10px">VI. Méthodes de Monte-Carlo à base de chaînes de Markov</h1>

<div class="alert alert-success"> Si $(X_n)$ est une chaîne de Markov de mesure stationnaire $f$, alors le théorème ergodique pour les chaînes de Markov garantit que
$$ \displaystyle \widehat I^{\text{MCMC}} = \frac{1}{N} \sum_{k=1}^N \phi(X_k) $$
est un estimateur convergent de $\int \phi f$: on a remplacé un échantillon i.i.d. par la trajectoire d'un processus de Markov, on parle alors de méthode <b>MCMC</b> (par opposition à IIDMC). Cette méthode n'a d'intérêt que si l'on sait générer une chaîne de Markov avec une probabilité stationnaire donnée : l'<b>algorithme de Metropolis-Hastings</b> fait précisément cela, il s'agit de la pierre angulaire des méthodes MCMC.</div>

<h2> VI.1. Algorithme de Metropolis–Hastings</h2>

L'algorithme de Metropolis--Hastings est un algorithme général qui permet de générer une chaîne de Markov avec une distribution invariante donnée. Soit $\pi$ une mesure de probabilité sur un ensemble dénombrable $\mathfrak{X}$ et $K: \mathfrak{X} \times \mathfrak{X} \to [0,1]$ un noyau de transition, *i.e.* pour tout $x \in \mathfrak{X}$ la famille $K(x, \cdot) = (K(x,y), y \in \mathfrak{X})$ définit une mesure de probabilité sur $\mathfrak{X}$. L'algorithme de Metropolis-Hastings construit la chaîne de Markov suivante :
1. Tirer $Y_t$ selon la loi $K(X_t, \, \cdot)$;
2. Calculer
$$ \displaystyle \varrho(X_t, Y_t) = \min \left( 1, \frac{\pi(Y_t)}{\pi(X_t)} \frac{K(Y_t, X_t)}{K(X_t, Y_t)} \right); $$
3. Choisir $\displaystyle X_{t+1} = \begin{cases}
    Y_t & \text{ avec probabilité } \varrho(X_t, Y_t),\\
    X_t & \text{ avec probabilité } 1 - \varrho(X_t, Y_t).
\end{cases} $

<p class="bg-primary" style="padding:1em"><b>QUESTION VI.1.1.</b> Calculez $p_{xy} = P(X_2 = y \mid X_1 = x)$ puis montrez que $\pi_x p_{xy} = \pi_y p_{yx}$. Déduisez-en que $X$ admet $\pi$ comme probabilité stationnaire.</p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION VI.1.1.</b> 

$$ p_{xy} = P(X_2 = y \mid X_1 = x) = P(tirer \ y \ étant \ à \ x) * P(accepter \ y \ étant \ à \ x \ et \ ayant \ tiré \ y)$$
$$ = K(x,y) \varrho(x, y) = K(x,y) \min \left( 1, \frac{\pi(y)}{\pi(x)} \frac{K(y, x)}{K(x, y)} \right)$$

Ainsi, $$ \pi_x p_{xy} = \pi_x K(x,y) \min \left( 1, \frac{\pi(y)}{\pi(x)} \frac{K(y, x)}{K(x, y)} \right) = \min \left( \pi_x K(x,y), \pi_y K(y,x) \right) = \pi_y K(y,x) \min \left( 1, \frac{\pi(x)}{\pi(y)} \frac{K(x, y)}{K(y, x)} \right) $$
Donc, $$ \pi_x p_{xy} = \pi_y p_{yx} $$

$\pi_x$ est la probabilité d'être à l'état x et $\pi_y$ est la probabilité d'être à l'état y. Donc, $\pi$ est la probabilité stationnaire de la chaîne.

</div>

<p class="bg-primary" style="padding:1em"><b>QUESTION VI.1.2.</b> Décrivez simplement cet algorithme lorsque le noyau de transition ne dépend pas de l'état de départ, <i>i.e.</i> $K(x, \cdot \,) = k$ pour une certaine mesure de probabilité $k$ et tout $x \in \mathfrak{X}$. Faites le lien avec la méthode du rejet.</p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION VI.1.2.</b> 

Dans ce cas, $\frac{K(y, x)}{K(x, y)} = 1$ partout, alors l'algorithme est le suivante:
1. Tirer $Y_t$ selon la loi k, indépendante de $X_t$;
2. Calculer
$$ \displaystyle \varrho = \min \left( 1, \frac{\pi(Y_t)}{\pi(X_t)} \right); $$
3. Tirer U selon une loi Uniforme [0,1]
4. Choisir $\displaystyle X_{t+1} = \begin{cases}
    Y_t & \text{ si U< $\varrho$ },\\
    X_t & \text{ sinon }.
\end{cases} $

Dans la méthode du rejet, on tire aussi U selon une loi uniforme, de parametres [0,k], et on accepte la valeur si $U \leq f(x)$, avec f la densité à simuler.

Dans notre cas, si $\pi(Y_t) > \pi(X_t)$ on accepte toujours la valeur. Cependant, si $\pi(Y_t) \leq \pi(X_t)$ on va fair un triage et accepter l'état $Y_t$ seulement si $U \leq \frac{\pi(Y_t)}{\pi(X_t)}$, ce qui est équivalent à accepter seulement les valeurs $Y_t$ telles que $U \leq \pi(Y_t)$, où U ~ Uniforme [0, $\pi(X_t)$] . On voit bien que ça ressemble à la méthode du rejet pour k = $\pi(X_t)$ et $f=\pi(Y_t)$, notre but avec l'algorithme de Metropolis-Hastings étant de créer une chaîne de Markov avec une distribution donnée.

</div>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION VI.1.4.</b> 

Tout d'abord, pour appliquer l'algortihme de Metropolis-Hastings, la connaissance de la distribuition  à une constante près suffit, puisque dans le rapport $\frac{\pi(Y_t)}{\pi(X_t)}$ avoir ou pas la valeur de la constante ne change pas la valeur. En outre, cet algorithme lié à Monte-Carlo, qui consiste dans la méthode MCMC (Monte-Carlo Markov Chaines), permet de résoudre de problèmes complexes en plusieurs dimensions où les autres méthodes étudiées échouent, et a une gamme d'applications très vaste et importante.

<h2> VI.2. Vitesse de convergence de la méthode MCMC</h2>

La méthode MCMC consiste donc à générer à l'aide de l'algorithme de Metropolis-Hastings une chaîne de Markov ayant une densité cible $f$ comme distribution stationnaire, puis de considérer l'estimateur
$$ \widehat I^{\text{MCMC}} = \frac{1}{N} \sum_{k=1}^N \phi(X_k) $$
qui est convergent par le théorème ergodique pour les chaînes de Markov.

<p class="bg-primary" style="padding:1em"><b>QUESTION VI.2.1.</b> A-t-on comme dans le cas IIDMC $\displaystyle \text{Var}(\widehat I^{\text{MCMC}}) = \frac{\text{Var}(\phi(X_1))}{N}$? Pourquoi?</p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION VI.2.1.</b> 

$$ \displaystyle \text{Var}(\widehat I^{\text{MCMC}}) = \text{Var}(\frac{1}{N} \sum_{k=1}^N \phi(X_k)) = \frac{1}{N^2}\text{Var}(\sum_{k=1}^N \phi(X_k)) $$

=$\frac{1}{N^2}(\sum_{k=1}^N \text{Var} \phi(X_k))$ seulement dans le cas où $\phi(X_k)$ indépendantes, ayant une covariance nulle, et $= \frac{1}{N}\text{Var} \phi(X_1)$ seulement au cas où $\phi(X_k)$ sont identiquement distribués.

Donc, on ne peut pas dire ça en général, puisque c'est une conséquence du cas où les $\phi(X_k)$ sont i.i.d.
</div>

<p class="bg-primary" style="padding:1em"><b>QUESTION VI.2.2.</b> Proposez une méthode numérique pour estimer $\text{Var}(\widehat I^{\text{MCMC}})$.</p>

Pour estimer $\text{Var}(\widehat I^{\text{MCMC}})$ on peut générer $c$ chaines de Markov ayant une densité cible $f$, générant $N$ états pour chaque chaîne. <h2> VI.3. Méthode MCMC pour l’estimation d'une intégrale sur la sphère</h2>

On revient à l'exemple de la sphère. On propose quatre noyaux de transition pour simuler une chaîne de Markov qui admette la distribution uniforme dans la sphère comme probabilité stationnaire :
* **Noyau uniforme - indépendant:** $K(x, \cdot \,)$ est la mesure uniforme sur $[-1,1]^d$;
* **Noyau uniforme - marche aléatoire:** $K(x, y) \propto {\bf 1}(\lVert y - x \rVert \leq \delta)$;
* **Noyau beta - indépendant:** $K(x, \cdot \,) = g_\theta$ avec $g_\theta$ introduit dans le BE sur l'échantillonnage préférentiel;
* **Noyau gaussien - marche aléatoire:** $K(x, y) \propto \exp \left( -\lVert y - x \rVert^2 / (2 \sigma^2) \right)$.

<p class="bg-primary" style="padding:1em"><b>QUESTION VI.3.1.</b> Décrivez par écrit l'algorithme de Metropolis-Hastings  dans chacun des ces cas.</p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION VI.3.1.</b> 

1. Tirer $Y_t$ selon la loi $k(x,y)$ de chaque cas. Dans les cas uniforme - indépendant et beta - indépendant, la loi de tirage de $Y_t$ ne dépend pas de $X_t$. Dans les cas uniforme - marche aléatoire et gaussien - marche aléatoire, on suit la même loi, toujours "centré" à $x$, l'argument de ces lois étant $\parallel y - x \parallel$. Ainsi, comme on a vue avant, pour ces lois on a $k(x,y) = k(\parallel y - x \parallel) = k(\parallel x - y \parallel) = k(y,x)$. Donc, quel que soit le cas, $\frac{K(y, x)}{K(x, y)} = 1$ partout, et comme on a aussi vu avant, il n'est pas nécessair de connaître les constantes de normalisation;
2. Calculer
$$ \displaystyle \varrho = \min \left( 1, \frac{\pi(Y_t)}{\pi(X_t)} \right); $$
Quel que soit le cas, on veut simuler une chaîne de Markov qui admette la distribution uniforme dans la sphère comme probabilité stationnaire. Donc, $ \frac{\pi(Y_t)}{\pi(X_t)}=1$ si $Y_t$ est dans la sphère et $=0$ si $Y_t$ déhors la sphère. Il faut donc évaluer si $Y_t$ est dédans ou déhors la sphère.
4. Choisir $\displaystyle X_{t+1} = \begin{cases}
    Y_t & \text{ si $Y_t$ dédans la sphère },\\
    X_t & \text{ sinon }.
\end{cases} $

</div>

L'objectif de cette section est d'utiliser la chaîne de Markov produite par l'algorithme de Metropolis-Hastings et le théorème ergodique pour estimer la valeur d'une intégrale sur la sphère,
$$ J = \int \phi(x) f_{\mathbb{S}}(x) dx $$
où $f_{\mathbb{S}}$ est la densité de la loi uniforme sur la sphère de rayon $1$ et $\phi(x) = \Vert x \Vert$ est la fonction à intégrer.

<p class="bg-primary" style="padding:1em"><b>QUESTION VI.3.2.</b> Complétez les fonctions <code>estimation_MCMC_XXX</code> qui implantent ces algorithmes de Metropolis-Hastings. Pour vérifier votre code, vous vérifierez numériquement et en petite dimension que ces quatre chaînes de Markov ont bien comme distribution stationnaire la mesure uniforme sur la sphère. Que vaut $J$? Comparez vos résultats avec la valeur attendue.</p>

```python
"""Dans les 4 cas, une fois que k(y,x)=K(x,y), et une fois que la distribution stationnaire cherchée est uniforme dans la
sphère, on va toujours accepter le tirage si Yt est dans la sphère et refuser si non."""

def estimation_MCMC_unif_ind(dim, t_max):
    Xt = np.zeros((t_max, dim))

    for i in range(t_max-1):
        Yt = np.random.uniform(low = -1, high = 1, size = dim)
        if(np.linalg.norm(Yt) <= 1):
            Xt[i+1,:] = Yt
        else:
            Xt[i+1,:] = Xt[i,:]
            
    norm = np.linalg.norm(Xt, axis =1)  
    estimation = np.sum(norm) / t_max
    J_th = dim / (dim+1)
    ecart_relatif = (estimation - J_th) / J_th
    return [estimation, ecart_relatif]

def estimation_MCMC_unif_MA(dim, t_max, delta):
    Xt = np.zeros((t_max, dim))

    for i in range(t_max-1):
        Yt = np.random.uniform(low = -delta, high = delta, size = dim)
        if(np.linalg.norm(Xt[i,:] + Yt) <= 1):
            Xt[i+1,:] = Xt[i,:] + Yt
        else:
            Xt[i+1,:] = Xt[i,:]
            
    norm = np.linalg.norm(Xt, axis =1) 
    estimation = np.sum(norm) / t_max
    J_th = dim / (dim+1)
    ecart_relatif = (estimation - J_th) / J_th
    return [estimation, ecart_relatif]
    
def estimation_MCMC_beta(dim, t_max, theta):
    Xt = np.zeros((t_max, dim))
    
    for i in range(t_max-1):
        Yt = np.random.beta(theta, theta, size = dim)
        Yt = 2 * Yt - 1
        if (np.linalg.norm(Yt) <= 1):
            Xt[i+1,:] = Yt
        else:
            Xt[i+1,:] = Xt[i,:]
            
    norm = np.linalg.norm(Xt, axis=1)
    estimation = np.sum(norm) / t_max
    J_th = dim / (dim+1)
    ecart_relatif = (estimation - J_th) / J_th
    return [estimation, ecart_relatif]
    

def estimation_MCMC_gaussien(dim, t_max):
    Xt = np.zeros((t_max, dim))
    
    for i in range(t_max-1):
        Yt = np.random.normal(Xt[i], 0.1, size = dim)
        if (np.linalg.norm(Yt) <= 1):
            Xt[i+1,:] = Yt
        else:
            Xt[i+1,:] = Xt[i,:]
            
    norm = np.linalg.norm(Xt, axis=1)
    estimation = np.sum(norm) / t_max
    J_th = dim / (dim+1)
    ecart_relatif = (estimation - J_th) / J_th
    return [estimation, ecart_relatif]

# Paramètres
dimensions = 2
t_max = 10**5
delta = 0.01
theta=1

# Calculs des valeurs
est_MCMC_unif_ind = estimation_MCMC_unif_ind(dimensions, t_max)
est_MCMC_unif_MA = estimation_MCMC_unif_MA(dimensions, t_max, delta)
est_MCMC_beta = estimation_MCMC_beta(dimensions, t_max, theta)
est_MCMC_gaussien = estimation_MCMC_gaussien(dimensions, t_max)
print("Valeur théorique =", dimensions/(dimensions+1))
print("MCMC_unif_ind : [estimation, écart relatif] =", est_MCMC_unif_ind)
print("MCMC_unif_MA  : [estimation, écart relatif] =",est_MCMC_unif_MA)
print("MCMC_beta     : [estimation, écart relatif] =",est_MCMC_beta)
print("MCMC_gaussien : [estimation, écart relatif] =",est_MCMC_gaussien)
```
