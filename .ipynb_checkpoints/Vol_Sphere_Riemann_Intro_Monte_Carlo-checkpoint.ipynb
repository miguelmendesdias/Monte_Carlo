```python
%matplotlib inline

import numpy as np
import sys
from scipy import special, stats
from scipy.optimize import minimize, fminbound
import random
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import pickle

matplotlib.rcParams['figure.figsize'] = (20,10)
matplotlib.rcParams['font.size'] = 22
matplotlib.rcParams['legend.handlelength'] = 2
matplotlib.rcParams['lines.linewidth'] = 3

def grille_Riemann(dim, taille_echantillon):
    """
    Retourne les centres d'hypercubes d'une grille multidimensionnelle
    """
    N = int(taille_echantillon**(1/dim)) # nb de boites par cote, eventuellement arrondi
    delta=2/N
    allG = [np.linspace(-1+delta/2,1-delta/2,num=N) for G in range(dim)]
    mesh = np.meshgrid(*allG)
    ans = list(zip(*(d.flat for d in mesh)))
    if len(ans) != taille_echantillon:
        print("  - Taille d'echantillon ajustee pour la methode de Riemann:"
              ,len(ans),"au lieu de",taille_echantillon,"(dimension =",dim,")")
    return np.array(ans)

def N_effectif(dim, taille_echantillon):
    """
    Retourne le nombre d'hypercubes utilisés lors de la discrétisation
    donne le même résultat que len(grille_Riemann(dim, taille_echantillon))
    mais sans générer la grille
    """
    return (int(taille_echantillon**(1/dim)))**dim
```

<a id="RIEMANN"></a><h1 style="border: 5px solid; text-align: center; padding: 10px"> I. Calcul du volume d'une sphère par la méthode de Riemann</h1>

<div class="alert alert-success">On montre dans cette partie les limitations des techniques d'intégration en grande dimension. L'exemple de l'estimation du volume d'une sphère sera le fil rouge du cours.</div>

Le volume $V_d$ délimité par la sphère de rayon $1$ en dimension $d$ est donné par l'expression suivante:

$$ V_d = \frac{\pi^{d/2}}{\Gamma(d/2+1)} \mathop{\sim}_{d \to \infty} \frac{1}{\sqrt{\pi d}} \left( \frac{2e\pi}{d} \right)^{d/2}
$$

où $\Gamma$ correspond à la fonction Gamma. En réexprimant $V_d$ sous forme intégrale,
$$ V_d = \int_{[-1,1]^d} f(x) dx \ \text{ avec } \ f(x) = {\bf 1}(\lVert x \rVert \leq 1) $$
on souhaite retrouver la valeur de $V_d$ en estimant l'intégrale ci-dessus à l'aide de sommes de Riemann. Si l'on divise $[-1,1]^d$ en hypercubes de côté $\delta$, on obtient alors l'approximation
$$ \int_{[-1,1]^d} f \approx \widehat V^{\text{R}}_d \ \text{ avec } \ \widehat V^{\text{R}}_d = \sum_{k=1}^N \delta^d f(x_k) = \frac{2^d}{N} \sum_{k=1}^N f(x_k) $$
où les $x_k$ sont les centres des hypercubes et $N  = (2/\delta)^d$ est le nombre d'hypercubes. Puisque $f$ est continue par morceaux, on a bien $\widehat V^{\text{R}}_d \to V_d$ lorsque le nombre $N$ d'hypercubes utilisés tend vers $+\infty$, *i.e.* lorsque le pas $\delta$ de la discrétisation tend vers $0$.

<div class="alert alert-success">On définit l'<b>écart relatif</b> comme la différence entre l'estimateur et sa valeur théorique, normalisée par la valeur théorique,
$$ \text{écart relatif} = \frac{ \widehat V - V }{V} $$
Ici, on connaît la valeur théorique et on peut donc calculer cet écart relatif, mais en pratique la valeur théorique, et donc l'écart relatif, est inconnue.</div>

<p class="bg-primary" style="padding:1em"><b>QUESTION I.1.</b> Complétez la fonction <code>estimation_Riemann</code> ci-dessous, puis utilisez le script suivant pour tracer sur une même courbe $\widehat V^{\text{R}}_d$ et sa valeur théorique lorsque $d$ varie entre $2$ et $13$, puis tracez l'évolution de l'écart relatif. Commentez les résultats obtenus et notamment l'influence de $N$ et de la dimension $d$. Expliquez en particulier le comportement de l'erreur lorsque la dimension grandit.</p>

```python
def estimation_Riemann(dim, grille):
    """
    Calcule l'approximation du volume de la sphère en dimension
    dim à l'aide de sommes de Riemann sur la grille fournie
    """
    somme = 0
    for xk in grille:
        if np.linalg.norm(xk) <= 1:
            somme +=1
    estimation = (2**dim) * somme / len(grille)
    V = (np.pi**(dim/2)) / special.gamma(dim/2 + 1)
    ecart_relatif = (estimation - V) / V
    return [estimation, ecart_relatif]
```

```python
# Paramètres
N = 10**6
dimensions = range(2,14)
nb_dimensions = len(dimensions)

# Calculs des valeurs
est_Riemann = np.zeros([nb_dimensions,2])
val_th = np.zeros(nb_dimensions)
for i in range(nb_dimensions):
    dim = dimensions[i]
    print("Estimation par la méthode de Riemann en dimension",dim)
    val_th[i] = (np.pi**(dim/2)) / special.gamma(dim/2 + 1)
    est_Riemann[i,:] = estimation_Riemann(dim, grille_Riemann(dim, N))

# Affichage
fig =plt.figure()
plt.plot(dimensions, est_Riemann[:,0], label="Riemann")
plt.plot(dimensions, val_th, 'r--', label="Théorie")
plt.legend()
fig.suptitle("Volume obtenu par Riemann vs Théorie")
plt.xlabel("Dimensions")
plt.show()

fig = plt.figure()
plt.plot(dimensions, est_Riemann[:,1])
fig.suptitle("Écart Relatif")
plt.xlabel("Dimensions")
plt.show()
```

<div class="alert alert-warning"><b>REPONSE A LA QUESTION I.1.</b> 

<p>On voit bien que pour des nombres de dimensions petits, la valeur obtenue par l'estimation de Riemann est presque la même que la valeur théorique, l'écart relatif étant presque nul. Donc, la méthode de Riemann est très bonne pour un nombre de dimensions petites. Cependant, quand le nombre de dimensions augmente, l'écart relatif augmente et les estimations deviennent pas bonnes. En effet, l'espace à remplir devient trop grand, et donc c'est difficile de bien le discretizer. Il faut remarquer que la taille d'échantillon diminue beaucoup lorsque le nombre de dimensions augmente, parce que quand le nombre de dimensions est grand, une petite augmentation du nombre d'hypercubes par axe se traduit par une grande augmentation du nombre d'hypercubes total.
    
Pour un nombre de dimensions fixé, si N augmente la qualité de l'estimation augmente, mais le temps de calcul augmente aussi.</p>

</div>

```python
vitesse_R = np.zeros(nb_dimensions)
for i in range(nb_dimensions):
    dim = dimensions[i]
    vitesse_R[i] = -np.log(np.abs(est_Riemann[i,0] - val_th[i])) / np.log(N_effectif(dim, N))
plt.plot(dimensions, vitesse_R)
plt.suptitle("Vitesse")
plt.xlabel("Dimensions")
plt.show()
```

<div class="alert alert-warning"><b>REPONSE A LA QUESTION I.2.</b> 

<p>Le code calcule la vitesse de convergence, à laquelle l'estimation de Riemann s'approche de la valeur théorique. Pour cela, on considere que $-log(|V_{estimée}-V_{théorique}|) = -log(|err|)$ est proportionnel à $log(N)$ selon un facteur $\alpha$ (notre vitesse de convergence). Cela est vrai si on assume que l'erreur diminue exponentiellement avec le nombre d'hypercubes N, c'est à dire $err=\exp(-\alpha log(N))$. Ainsi, pour un certain N, l'erreur diminue lorsque alpha augmente. Alors, la vitesse de convergence est plus élevée si le nombre de dimensions est petit. Il faut noter que le code utilise bien $N_{effectif}$, qui est le nombre d'hypercubes effectivement utilisé, et pas le N choisi de $10^6$. On voit bien que pour des dimensions $\geq 10$ la convergence est très mauvaise.</p>

</div>

On s'intéresse à la robustesse de l'estimation précédente. Pour cela, on considère
$$ \widehat V^{\text{Rc}}_d = \frac{2^d}{N} \sum_{k=1}^N f(c_k) $$
où $c_k$ est le coin de l'hypercube $k$, i.e., $c_k = x_k + \frac{\delta}{2} {\bf 1}$ avec ${\bf 1} = (1, \ldots, 1) \in R^n$.

<p class="bg-primary" style="padding:1em"><b>QUESTION I.3.</b> Définissez la fonction <code>estimation_Riemann_coin</code> afin qu'elle renvoie l'estimation et l'écart correspondant à cette nouvelle approximation. Rajoutez sur les courbes précédentes les résultats liés à cette estimation et commentez les résultats obtenus.</p>

```python
def estimation_Riemann_coin(dim, grille):
    """
    Calcule l'approximation du volume de la sphère en dimension
    dim à l'aide de sommes de Riemann par la méthode coin sur la grille fournie
    """
    somme = 0
    delta = (2) / (len(grille)**(1/dim))
    grille += delta/2
    for xk in grille:
        if np.linalg.norm(xk) <= 1:
            somme +=1
    estimation = (2**dim) * somme / len(grille)
    V = (np.pi**(dim/2)) / special.gamma(dim/2 + 1)
    ecart_relatif = (estimation - V) / V
    return [estimation, ecart_relatif]
```

```python
# Calculs des valeurs
est_Riemann = np.zeros([nb_dimensions,2])
est_Riemann_coin = np.zeros([nb_dimensions,2])
for i in range(nb_dimensions):
    dim = dimensions[i]
    print("Estimation par la méthode de Riemann en dimension",dim)
    val_th[i] = (np.pi**(dim/2)) / special.gamma(dim/2 + 1)
    est_Riemann[i,:] = estimation_Riemann(dim, grille_Riemann(dim, N))
    est_Riemann_coin[i,:] = estimation_Riemann_coin(dim, grille_Riemann(dim,N))

# Affichage
fig = plt.figure()
plt.plot(dimensions, est_Riemann[:,0], label="Riemann")
plt.plot(dimensions, est_Riemann_coin[:,0], label="Riemann coin")
plt.plot(dimensions, val_th, 'r--', label="Théorie")
plt.legend()
fig.suptitle("Volume calculé par les méthodes de Riemann vs Volume Théorique")
plt.xlabel("Dimensions")

fig = plt.figure()
plt.plot(dimensions, est_Riemann[:,1], label="Riemann")
plt.plot(dimensions, est_Riemann_coin[:,1], label="Riemann coin")
plt.legend()
fig.suptitle("Écarts Relatifs")
plt.xlabel("Dimensions")
```

<div class="alert alert-warning"><b>REPONSE A LA QUESTION I.3.</b> 

<p>On voit que le choix de décaller un peut les centres des hypercubes ne change pas beaucoup les résultats obtenus, même s'ils sont légèrement différents. La méthode devient instable quand le nombre de dimensions augmente et n'est pas bonne pour ces cas.</p>

<p class="bg-primary" style="padding:1em"><b>QUESTION I.4.</b> Soit $V^-_d$ le volume de la sphère de rayon $1 - K/d$ en dimension $d$ : montrez théoriquement que $V_d - V^-_d \sim (1-e^{-K}) V_d$ lorsque $d \to \infty$, et utilisez ce résultat pour interpréter les résultats précédents.</p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION I.4.</b> 

Si $V_d$ est le volume de la sphère de rayon 1, alors le volume de la sphère de rayon  $1 - K/d$ en $d$ dimensions est donné par $V^-_d = V_d (1 - \frac{K}{d})^d$.

Ainsi, $$V_d - V^-_d = V_d (1-(1 - \frac{K}{d})^d)$$

On calcule la limite lorsque $d \rightarrow \infty,$

$$ lim_{d \rightarrow \infty} V_d [1 - (1 - \frac{K}{d})^d] = V_d [1 - lim_{d \rightarrow \infty}(1 - \frac{K}{d})^d]
=V_d [1 - lim_{d \rightarrow \infty}(1 + \frac{1}{\frac{-d}{K}})^d] = V_d [1 - [lim_{d \rightarrow \infty}(1 + \frac{1}{\frac{-d}{K}})^{-d/K}]^{-K}] $$

$$= V_d (1 - e^{-K}) $$

On voit que si $d \rightarrow \infty$, K peut augmenter autant qu'on veut et le rayon sera toujours une valeur de presque 1. Cependant, même si la valeur du rayon reste presque constante, si K augmente beaucoup, $V^-_d = V_d (e^{-K}) \rightarrow 0$. Ainsi, la sphère de rayon 1 est infiniment plus grande qu'une sphère de rayon inférieur à 1 mais aussi proche de 1 qu'on veut. Cela veut dire que quand $d \rightarrow \infty$, le volume de la sphère est tout concentré dans sa surface.

Quand on fait une grille de Riemann, on distribue les hypercubes de façon homogène dans l'espace $[-1,1]^d$, ce qu'explique les valeur pas bonnes obtenues avec cette méthode pour des dimensions élevées. En effet, la majorité des hypercubes ne servent à rien, étant presque tous en déhors de la sphère.

La méthode Riemann coin pourrait améliorer un peu les résultats, une fois qu'il y a une simple décallage des centres des hypercubes, mais on a en fait vu que cette méthode donne des résultats pires que la méthode de Riemann traditionnelle pour des nombres de dimensions élevés.

</div>

<p class="bg-primary" style="padding:1em"> <b>$\star$ QUESTION I.5 $\star$.</b> Quel bilan tirez-vous de l'estimation d'intégrales par somme de Riemann? </p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION I.5.</b> 

Pour des nombres de dimensions petits, la méthode de Riemann permet d'arriver à un bon résultat, qui est bien sûr mieux pour des dimensions plus petites. Le résultat est plus précis si la taille d'échantillonage $N$ augmente, mais les calcus vont prendre plus de temps.

Au fur et à mesure que le nombre de dimensions augmente, plusieurs problèmes apparaîssent. D'abbord, la maille devient moins fine. En effet, si on maintien $N$ pour $d$ plus élevé, alors ayant le même nombre d'hypercubes à distribuer pour plus de dimensions, $\delta$ va augmenter, et la discretization de chaque axe devient pire. En plus, le $N_{effectif}$ va en principe diminuer beaucoup.

En outre, nous avons vu que quand $d \rightarrow \infty$ tout le volume de la sphère se concentre dans sa surface. La grille de Riemann étant uniformément distribué, la majorité des hypercubes ne vont servir à rien, puisqu'ils sont presque tous en déhors de la sphère. Ce serait mieux d'avoir plus de points dans la surface, qui donne la grande contribution pour le volume.

On peut dire le même pour n'importe quelle intégrale qu'on veut calculer: il faut avoir plus de points dans les régions où l'intégrale est plus grande.

Donc, le grand problème de la méthode de Riemann se trouve dans le placement des points dans une grille, laquelle n'est pas adapté à chaque problème spécifique qu'on puisse avoir.

<a id="MC"></a><h1 style="border: 5px solid; text-align: center; padding: 10px">II. Introduction à la méthode de Monte-Carlo : le cas IID</h1>

<div class="alert alert-success">On introduit maintenant la méthode de Monte-Carlo sur le même problème d'estimation du volume de la sphère et on discute sa vitesse de convergence.<br>
<b>Rappel</b> i.i.d. = indépendantes et identiquement distribuées</div>

<h2>II.1. Méthode de Monte-Carlo</h2>

Soit $(X_k, k \geq 1)$ des variables i.i.d. uniformément réparties sur $[-1,1]^d$ et
$$ \widehat V^{\text{MC}}_d = \frac{2^d}{N} \sum_{k=1}^N f(X_k). $$

<p class="bg-primary" style="padding:1em"><b>QUESTION II.1.1.</b> Donnez la densité de $X_1$ et justifiez que $\widehat V^{\text{MC}}_d \to V_d$ lorsque $N \to \infty$.</p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION II.1.1.</b> 

Vue que $(X_k)$ sont uniformément réparties sur $[-1,1]^d$, ces fonctions densité de probabilité sont données par:

$$ g_{X_k}(x) = g_{X_1}(x) =\frac{1}{2^d} \mathbb{1}_{[-1,1]^d} $$
Par définition,
$$ V_d = \int \mathbb{1}(\parallel x \parallel \leq 1) dx = \int 2^d \frac{\mathbb{1}(\parallel x \parallel \leq 1)}{2^d} dx
       = 2^d E[\mathbb{1}(\parallel X \parallel \leq 1)], $$
       
avec $X \mathop{\sim}Uniforme({[-1,1]^d}).$
       
D'après la Loi des Grands Nombres, lorsque $N \to \infty$, un estimateur de cette valeur est donné par:

$$ 2^d \frac{1}{N} \sum_{k=1}^N \mathbb{1}(\parallel X_k \parallel \leq 1) = \frac{2^d}{N} \sum_{k=1}^N f(X_k) = \widehat V^{\text{MC}}_d,$$
avec $X_k \mathop{\sim}Uniforme({[-1,1]^d})$ et $f(X_k)=\mathbb{1}(\parallel X_k \parallel \leq 1) $

Donc, d'après la Loi des Grands Nombres,$$lim_{N \to \infty} \widehat V^{\text{MC}}_d = V_d$$.

</div>

<p class="bg-primary" style="padding:1em"><b>QUESTION II.1.2.</b> Complétez la fonction <code>estimation_IIDMC</code> afin d'ajouter aux courbes précédentes l'évolution de $\widehat V^{\text{MC}}_d$ ainsi que l'écart relatif correspondant. Commentez les courbes obtenues.</p>

```python
def estimation_IIDMC(dim, taille_echantillon):
    """
    Calcule l'approximation du volume de la sphére en dimension
    dim à l'aide de la méthode de Monte-Carlo
    """
    somme = 0
    for i in range(taille_echantillon):
        rand_vector = np.random.uniform(low = -1, high = 1, size = dim)
        if np.linalg.norm(rand_vector) <= 1:
            somme += 1
    # pour pouvoir comparer, on utilise N_effectif puisque c'est cette valeur que la méthode de Riemann a utilisé opur la dimension en question
    estimation = (2**dim) / taille_echantillon * somme
    V = (np.pi**(dim/2)) / special.gamma(dim/2 + 1)
    ecart_relatif = (estimation - V) / V
    erreur_relative = ((((2.0**dim) / V) - 1.0) / taille_echantillon)**(.5)
    return [estimation, ecart_relatif, erreur_relative]
```

```python
# Calculs des valeurs     erreur_relative = cd / np.sqrt(taille_echantillon)
est_IIDMC = np.zeros([nb_dimensions,3])
for i in range(nb_dimensions):
    dim = dimensions[i]
    print("Estimation par Monte Carlo en dimension:",dim)
    # pour une comparaison juste, on prend la même taille d'échantillon que pour Riemann
    est_IIDMC[i,:] = estimation_IIDMC(dim, N_effectif(dim, N))
    print("[Estimation, Écart Relatif] = [", est_IIDMC[i,0], ",", est_IIDMC[i,1], "]")

# Affichage
fig = plt.figure()
plt.plot(dimensions, est_IIDMC[:,0], label="IIDMC")
plt.plot(dimensions, est_Riemann[:,0], label="Riemann")
plt.plot(dimensions, est_Riemann_coin[:,0], label="Riemann coin")
th, = plt.plot(dimensions, val_th, 'r--', label="Théorie")
plt.legend()
fig.suptitle("Estimations de Volume de la sphère vs Valeur Théorique")
plt.xlabel("Dimensions")

fig = plt.figure()
plt.plot(dimensions, est_IIDMC[:,1], label="IIDMC")
plt.plot(dimensions, est_Riemann[:,1], label="Riemann")
plt.plot(dimensions, est_Riemann_coin[:,1], label="Riemann coin")
plt.legend()
fig.suptitle("Écarts Relatifs")
plt.xlabel("Dimensions")
```

<div class="alert alert-warning"><b>REPONSE A LA QUESTION II.1.2.</b> 

On peut constater que les estimations utilisant la méthode IIDMC sont beaucoup meilleures que les estimations obtenues avec la méthode de Riemann lorsque le nombre de dimensions augmente. En effet, pour un nombre de dimensions inférieur ou égal à 13, l'écart relatif reste très proche de 0. On a utilisé une taille d'échantillon égale à $N_{effectif}$ utilisé lors de l'utilisation de la méthode de Riemann pour pouvoir comparer les méthodes, mais si on avait utilisé la vraie taille demandé les valeurs avec Monte-Carlo seraient encore meilleures.

<p class="bg-primary" style="padding:1em"><b>QUESTION II.1.4.</b> Quelle est la vitesse de convergence théorique de l'estimateur $\widehat V^{MC}_d$? Cette vitesse dépend-elle de la dimension ? Commentez par rapport aux résultats numériques précédents.</p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION II.1.4.</b> 

On a vu que si $(X_k)$ sont uniformément réparties sur $[-1,1]^d$, alors
$$ E[\widehat V^{MC}_d] = \mu = V_d = 2^d E[\mathbb{1}(\parallel X_k \parallel \leq 1)] $$

On a aussi: $$\sigma^2 = Var[V_d] = 2^{2d} Var[\mathbb{1}(\parallel X_k \parallel \leq 1)] $$

Puis, d'après le Théoreme Central Limite, si $Y_k \sim i.i.d$:

$$lim_{N \to \infty} \frac{\sum_{k=1}^N (Y_k - \mu)}{\sqrt{N \sigma^2}} \mathop{\sim} N(0,1) \Leftrightarrow \frac{1}{N}  lim_{N \to \infty} {\sum_{k=1}^N (Y_k - \mu)} \mathop{\sim} N(0,\frac{\sigma^2}{N})$$

Enfin, si on applique cela à notre cas, où $Y_k = 2^d \mathbb{1}(\parallel X_k \parallel \leq 1)$, on obtient:

$$\widehat V^{MC}_d - V_d \mathop{\sim} N(0,(\frac{\sigma}{\sqrt N})^2) $$

Donc, si l'on considère que la vitesse de convergence de l'estimateur est proportionnelle à l'inverse de l'ecart-type de l'estimateur, la vitesse de convergence est alors proportionnel à $\sqrt N$. Cela veut dire que la vitesse de convergence augmente lorsque la taille d'échantillonage augmente.

La vitesse de convergence dépend aussi du nombre de dimensions, une fois qu'elle varie inversement à $\sigma$. En effet, lorsque $d$ augmente, on voit que $\sigma^2 = 2^{2d} Var[\mathbb{1}(\parallel X_k \parallel \leq 1)] $ augmente, et donc la vitesse de convergence diminue.

En ce qui concerne les résultats numériques obtenus, ils montrent une évolution de la vitesse de convergence avec des oscillations, alors qu'on vient de dire que la vitesse de convergence devrait diminuer lorsque $d$ augmente. On n'a toutefois pas utilisé toujours la même valeur de N (on a utilisé la valeur $N_{effectif}$), ce qui peut expliquer les résultats obtenus. En effet, N influence aussi la vitesse de convergence. De plus, on travail avec des variables aléatoires, donc on n'obtient pas toujours les mêmes résultats.

À chaque itération du calcul les valeurs obtenues sont différentes, ce qui était attendue, puisque l'estimateur est lui même une variable aléatoire. En revanche, on voit que l'estimation, ainsi que la vitesse de convergence, sont stables et bornées.

<h2>II.2. Précision d'un estimateur</h2>

Le but de cette partie est de montrer que la variance n'est toujours pas une mesure satisfaisante pour mesurer la précision d'un estimateur. On considère pour cela la fonction $p(t) = P(\lVert X \rVert \leq t)$ où $X$ est uniformément répartie sur $[-1,1]^d$.

<p class="bg-primary" style="padding:1em"><b>QUESTION II.2.1</b> Donnez l'estimateur de Monte-Carlo de $p(t)$, que l'on notera $\widehat p(t)$, ainsi que son espérance et sa variance. En déduire que $\frac{1}{N} (\widehat p(t) - \widehat p(t)^2)$ est un estimateur de la variance de $\widehat p(t)$.</p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION II.2.1.</b> 

$$ p(t) = P(\lVert X \rVert \leq t) = \int_{[-1,1]^d} \mathbb{1}(\parallel x \parallel \leq t) \frac{1}{2^d} dx = E[\mathbb{1}(\parallel X \parallel \leq t)] $$
Ainsi, d'après la Loi des Grands Nombres, l'estimateur de Monte Carlo IIDMC est donné par:

$$ \widehat p(t) = \frac{1}{N} \sum_{k=1}^N \mathbb{1}(\parallel X_k \parallel \leq t), X_t \sim U_{[-1,1]^d} $$

De la façon qu'on a construit l'estimateur, l'espérance est évidemment $V_d$, la valeur qu'on cherche à estimer. En effet, si $X_k$ sont i.i.d, alors

$$ E[\widehat p(t)] = E[\frac{1}{N} \sum_{k=1}^N \mathbb{1}(\parallel X_k \parallel \leq t)] = E[\mathbb{1}(\parallel X_1 \parallel \leq t)] = p$$
L'estimateur est donc sans biais et convergent. $X_k$ étant i.i.d, les covariances sont nulles et Var de la somme est la somme des Var. Donc,

$$Var[\widehat p(t)] = Var[\frac{1}{N} \sum_{k=1}^N \mathbb{1}(\parallel X_t \parallel \leq t)] = \frac{1}{N^2} \sum_{k=1}^N Var[\mathbb{1}(\parallel X_k \parallel \leq t)] =  \frac{1}{N} Var[\mathbb{1}(\parallel X_k \parallel \leq t)] $$

Une fois que $\mathbb{1}(\parallel X_k \parallel \leq t)$ vaut soit 0 soit 1, $\mathbb{1}(\parallel X_k \parallel \leq t)^2=\mathbb{1}(\parallel X_k \parallel \leq t)$ et on a:

$$ Var[\mathbb{1}(\parallel X_t \parallel \leq t)] = E[[\mathbb{1}(\parallel X_k \parallel \leq t)]^2] - [E[\mathbb{1}(\parallel X_k \parallel \leq t)]]^2 = p(t) - [p(t)]^2 $$

Pour finir, si on utilise au lieu de p son estimateur $ \widehat p(t) $:

$$\widehat{Var[\widehat p(t)]} = \frac {\widehat p(t) - \widehat p(t)^2}{N}$$

</div>

<p class="bg-primary" style="padding:1em"><b>QUESTION II.2.2.</b> Complétez le script suivant afin de tracer $t \mapsto \widehat p(t) - \widehat p(t)^2$. Pour quelle valeur de $t$ l'estimateur vous semble-t-il être le moins précis ?</p>

```python
taille_echantillon = 10**4
dim = 15
echantillon = np.random.uniform(low=-1, high=1, size=(taille_echantillon, dim))
echantillon_norme = np.zeros(taille_echantillon)
for i in range(taille_echantillon):
    echantillon_norme[i] = np.linalg.norm(echantillon[i,:])
valeurs_t = np.arange(0,3,.01)
estimation_p = np.zeros(len(valeurs_t))
for i in range(len(valeurs_t)):
    somme = 0
    for norme in echantillon_norme:
        if norme <= valeurs_t[i]:
            somme += 1
    estimation_p[i] = somme / taille_echantillon
    

plt.plot(valeurs_t, estimation_p - np.power(estimation_p,2), label="Var(estimation)")
plt.plot(valeurs_t, estimation_p, label="Estimation")
plt.legend()
plt.xlabel("t")
plt.suptitle("Estimation et sa variance")
```

<div class="alert alert-warning"><b>REPONSE A LA QUESTION II.2.2.</b> 

Anlysant le graphique pour déterminer la valeur de t correspondant à un estimateur moins précis, on pourrait penser que ce serait la valeur de t correspondant à la variance maximale, une fois que la variance mesure la disperson des échantillons. Dans ce cas, la valeur serait de $t \approx 2.25$. Néanmoins, si on réfléchit, il ne fait pas du sens comparer la dispersion absolute sans considérer la valeur à atteindre. En effet, supposons une estimation d'une valeur de $10^{-2}$ avec une variance de $10^{-2}$ est une autre estimation de 1000 avec une variance de 0.1. La dernière présente une variance plus élevée même si elle est plus précis en termes relatifs.

Donc, l'estimation de la variance seule n'est pas un bon critère pour déterminer la précision de l'estimateur. 

</div>

<p class="bg-primary" style="padding:1em"><b>QUESTION II.2.3.</b> Justifiez le choix de
$$ \frac{\sqrt{\widehat p(t) - \widehat p(t)^2}}{\widehat p(t)} $$
pour évaluer la précision de $\widehat p(t)$.</p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION II.2.3.</b> 

Comme on vient de dire, l'estimation de la variance seule n'est pas un bon critère pour déterminer la précision de l'estimateur. En effet, supposons une estimation d'une valeur de $10^{-2}$ avec une variance de $10^{-2}$ est une autre estimation de 1000 avec une variance de 0.1. La dernière présente une variance plus élevée même si elle est plus précis en termes relatifs. L'écart type est en une mesure de l'erreur absolue. Donc, une suggestion pour évaluer la précision de l'estimateur serait de diviser son écart-type par sa espérance, c'est à dire $\frac{\widehat {\sigma(\widehat p(t))}}{\widehat p(t)} = \frac{1}{\sqrt{N}}\frac{\sqrt{\widehat p(t) - \widehat p(t)^2}}{\widehat p(t)} $.

Néanmoins, pour enlever la dépendence de N, et obtenir la dépendance de la précision avec le parametre t seulement, pour un N fixé, on peut prendre comme mesure:

$$ \frac{\sqrt{\widehat p(t) - \widehat p(t)^2}}{\widehat p(t)} $$

</div>

<div class="alert alert-success center;">On définit l'<b>erreur relative</b> comme le rapport entre l'écart-type de l'estimateur et sa moyenne :
$$ \text{erreur relative} = \frac{\sqrt{\text{Var}(\widehat V)}}{E (\widehat V)} $$
En pratique, ces valeurs sont inconnues mais on peut les estimer par Monte-Carlo.</div>

Quand t augmente le contraire se passe, et c'est pourquoi la précision augmente.

<h2>II.3. Limitations de la méthode IIDMC</h2>

On conclut cette partie par une limitation évidente de la méthode de Monte-Carlo.

<p class="bg-primary" style="padding:1em"><b>QUESTION II.3.1.</b> Que vaut $\widehat V^{\text{MC}}_{30}$ pour $N = 10^6$? Faites un calcul d'ordre de grandeur pour déterminer le nombre d'échantillons qu'il faudrait pour avoir une erreur relative $\leq 10^{-2}$.</p>

```python
N = 10**6
dim = 30
V = (np.pi**(dim/2)) / special.gamma(dim/2 + 1)
print(estimation_IIDMC(dim, N))
N_pour_err_rel_souhaitee = (2**dim / V -1) / (10**(-4))
print("Pour obtenir l'erreur relative souhaitée il faut N =", N_pour_err_rel_souhaitee)
```

<div class="alert alert-warning"><b>REPONSE A LA QUESTION II.3.1.</b> 

Pour obtenir une erreur relative de $10^{-2}$ il faudrait que:

$$ 10^{-2} = \frac{c_d}{\sqrt N} = \frac{\sqrt{\frac{2^d}{V_d}-1}}{\sqrt N} $$

Donc, pour dim=30 il faut:

$$ N = \frac{\frac{2^{30}}{V_{30}}-1}{10^{-4}} \sim 10^{17}$$

On conclut que pour 30 dimensions, pour obtenir l'erreur relative souhaitée, N est trop grand, et donc le temps de calcul est trop grand et la méthode IIDMC n'est pas faisable.
</div>

<p class="bg-primary" style="padding:1em"><b>$\star$ QUESTION II.3.2. $\star$</b> Quel bilan tirez-vous de l'estimation d'intégrales par méthode de Monte-Carlo? </p>

<div class="alert alert-warning"><b>REPONSE A LA QUESTION II.3.2.</b> 

En bref, on a vu que les méthodes de Riemann permettent de calculer l'intégrale souhaitée avec une bonne précision pour un nombre de dimensions petit. Par contre, quand le nombre de dimensions augmente, la méthode de Riemann n'est plus bonne. Cela est dû au fait que la méthode de Riemann utilise une grille de points uniformément réparitis dans l'espace. En effet, en faisant ça, la grille est général et pas bien adapté au cas spécifique qu'on traite. Par exemple, dans le cas du volume de la sphère, lorsque $d \to \infty$, presque aucun hypercube n'est dans la sphère, et donc l'estimation devient très mauvais.

En revanche, la méthode de Monte Carlo a permis d'obtenir beaucoup meilleurs résultats que Riemann pour des nombres de dimensions supérieures (restant bonne pour des nombres de dimensions inférieures). Cela faisant jusqu'à maintenant que des tirages aléatoires uniformes, c'est à dire même ne prennant encore en compte quelle est la fonction que l'on veut intégrer. On va voir par la suite qu'on peut encore utiliser d'autres méthodes pour améliorer les estimations. La méthode de Monte Carlo est donc beaucoup mieux que la méthode de Riemann lorsque le nombre de dimensions augmente. C'est quand même vrai que sa vitesse de convergence diminue lorsque le nombre de dimensions augmente. C'est aussi vrai que l'erreur relatif augmente quand le nombre de dimensions augmente. Enfin, la vitesse de convergence augmente avec $\sqrt N$.

</div>

<div class="alert alert-danger"><b>On a illustré ci-dessus la méthode de Monte-Carlo sur l'exemple du calcul de l'intégrale
$$ \int_{[-1,1]^d} \phi(x) d x \ {\bf {\text{ avec }}} \ \phi(x) = {\bf 1}(\lVert x \rVert \leq 1). $$
L'idée a été d'interpréter l'intégrale comme une espérance et d'utiliser la loi forte des grands nombres pour approcher cette espérance. Cette idée se généralise aisément au calcul de l'intégrale
$$ \int \phi(x) f(x) d x $$
où $f$ est une densité, en réécrivant cette intégrale comme l'espérance de la variable aléatoire $\phi(X)$ où $X$ est de densité $f$. Par ailleurs, contrairement au cas déterministe, la vitesse de convergence de cette méthode est indépendante de la dimension du problème!</b></div>
